## Data Transformation

###Create new columns based on existing ones

#### Confirm Data Types

```python
# Confirm data types
datatypes = healthcare_cleaned.dtypes
print(datatypes)
```

#### Add a new column to designate Normal or High Average Heart Rates

```python
# Convert AverageHeartRate column to int
healthcare_cleaned['AverageHeartRate'] = healthcare_cleaned['AverageHeartRate'].apply(int)

# Create threshold value
threshold_high = 100

# Create a new column "HeartRateStatus" based on thresholds
healthcare_cleaned['HeartRateStatus'] = healthcare_cleaned['AverageHeartRate'].apply(lambda x: 'High' if x >= threshold_high else 'Normal')

# Confirm new column is added
print(healthcare_cleaned)
```

#### Add a new column to designate if the visit diagnoses was a viral illness

```python
# Define criteria for classifying diseases as "Viral"
viral_diseases = ['Covid-19', 'Flu', "Cold"]

# Create a new column "ViralDiagnosis" based on the criteria
healthcare_cleaned['ViralDiagnosis'] = healthcare_cleaned['DiseaseDiagnosed'].apply(lambda x: 'Viral' if x in viral_diseases else 'Non-Viral')

# Confirm new column is added
print(healthcare_cleaned)
```

### Aggregate data using groupby and compute summary statistics

#### Group average heart rates by patient age

```python
grouped_hr = healthcare_cleaned.groupby('PatientAge').agg({
    'AverageHeartRate':'mean'
})
print(grouped_hr)
```

#### Group average prescription cost by diagnosed disease

```python
# Convert PrescriptionCostInDollars column to float
healthcare_cleaned['PrescriptionCostInDollars'] = healthcare_cleaned['PrescriptionCostInDollars'].apply(float)

grouped_cost = healthcare_cleaned.groupby('DiseaseDiagnosed').agg({
    'PrescriptionCostInDollars':'mean'
})
print(grouped_cost)
```

```python
# Sum average prescription cost by Disease Diagnosed
grouped_costSum = healthcare_cleaned.groupby('DiseaseDiagnosed').agg({
    'PrescriptionCostInDollars':'sum'
})
print(grouped_costSum)
```

### Use pivot tables or cross-tabulations for multi-dimensional analysis

#### Pivot Tablet: Average Visit Duration in Minutes by Number of Tests per Visit



```python
# Convert PrescriptionCostInDollars column to float
healthcare_cleaned['VisitDurationMins'] = healthcare_cleaned['VisitDurationMins'].apply(int)

healthcare_cleaned['VisitLength'] = pd.cut(healthcare_cleaned['VisitDurationMins'], bins=[0, 30, 60, 120, 240, np.inf], labels=['0-30', '30-60', '60-120', '120-240', 'over 240'])
pivot_data = healthcare_cleaned.pivot_table(columns='VisitLength', index='NumberOfTests', values='VisitDurationMins', aggfunc='mean')
print(pivot_data)

```

#### Pivot Table: Average Prescription Cost by Patient Insurance Status

```python
healthcare_cleaned['RXCostInDollars'] = pd.cut(healthcare_cleaned['PrescriptionCostInDollars'], bins=[0, 20, 40, 60, 80, 100, 200, np.inf], labels=['0-20', '20-40', '40-60', '60-80', '80-100', '100-200', 'over 200'])
pivot_data1 = healthcare_cleaned.pivot_table(columns='RXCostInDollars', index='HasInsurance', values='PrescriptionCostInDollars', aggfunc='mean')
print(pivot_data1)
```
